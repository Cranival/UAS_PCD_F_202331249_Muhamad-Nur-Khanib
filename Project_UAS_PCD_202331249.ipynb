# Filtering Citra

```python
img_original = cv2.imread('filtering_citra.jpg')
#202331249_Muhamad Nur Khanib
```

### Filter Median

```python
img_original_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB) # Untuk tampilan median agar berwarna
citra_grayscale = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY) # Untuk tampilan original grayscale & input mean filter

# Terapkan filter median pada citra RGB
img_median_after = cv2.medianBlur(img_original_rgb, 5)
#202331249_Muhamad Nur Khanib
```

### Filter Mean

```python
#202331249_Muhamad Nur Khanib
# Gunakan citra_grayscale dari Note 1 sebagai input
copyCitra1 = citra_grayscale.copy().astype(float)

m1, n1 = copyCitra1.shape
output_mean_manual = np.empty([m1, n1])

# Loop untuk filter rata-rata 3x3
for baris in range(1, m1 - 1):
    for kolom in range(1, n1 - 1):
        jumlah = (copyCitra1[baris-1, kolom-1] + copyCitra1[baris-1, kolom] + copyCitra1[baris-1, kolom+1] +
                  copyCitra1[baris, kolom-1]   + copyCitra1[baris, kolom]   + copyCitra1[baris, kolom+1] +
                  copyCitra1[baris+1, kolom-1] + copyCitra1[baris+1, kolom] + copyCitra1[baris+1, kolom+1])
        output_mean_manual[baris, kolom] = (1/9) * jumlah
```

### Hasil filter

```python
#202331249_Muhamad Nur Khanib
plt.figure(figsize=(15, 10))

# Gambar 1 (atas kiri): Gambar original grayscale
plt.subplot(2, 2, 1)
plt.imshow(citra_grayscale, cmap='gray')
plt.title("Citra Asli (Grayscale)")
plt.xticks([]), plt.yticks([])

# Gambar 2 (atas kanan): Gambar setelah filter median (berwarna)
plt.subplot(2, 2, 2)
plt.imshow(img_median_after) # img_median_after sudah dalam format RGB
plt.title("Setelah Filter Median")
plt.xticks([]), plt.yticks([])

# Gambar 3 (bawah kiri): Gambar original grayscale 
plt.subplot(2, 2, 3)
plt.imshow(citra_grayscale, cmap='gray')
plt.title("Citra Asli (Grayscale)")
plt.xticks([]), plt.yticks([])

# Gambar 4 (bawah kanan): Gambar setelah filter rata-rata (grayscale)
plt.subplot(2, 2, 4)
plt.imshow(output_mean_manual, cmap='gray')
plt.title("Setelah Filter Rata-rata (Manual)")
plt.xticks([]), plt.yticks([])

plt.tight_layout() 
plt.show()
```

# Geometriks Citra 

```python
#202331249_Muhamad Nur Khanib
img = cv2.imread('geometriks_citra.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows, cols, _ = img_rgb.shape
```

### Transformasi Geometris (Rotasi, Ubah Ukuran, Potong, Balik, Geser)

```python
# 202331249_Muhamad Nur Khanib
# --- 1. Rotasi (Rotated) ---
# Memutar citra 25 derajat searah jarum jam di sekitar pusatnya
M_rot = cv2.getRotationMatrix2D(((cols-1)/2.0, (rows-1)/2.0), 25, 1)
img_rotated = cv2.warpAffine(img_rgb, M_rot, (cols, rows))

# --- 2. Ubah Ukuran (Resized) ---
# Mengubah ukuran citra dari dimensi aslinya
img_resized = cv2.resize(img_rgb, None, fx=0.3, fy=0.5, interpolation=cv2.INTER_AREA)

# --- 3. Potong (Cropped) ---
# Memotong sebagian citra.
# [y_start:y_end, x_start:x_end]
# Ini akan memotong area sekitar mata
img_cropped = img_rgb[300:600, 900:1080]

# --- 4. Balik (Flipped) ---
# Membalik citra secara horizontal (flipCode = 1)
img_flipped = cv2.flip(img_rgb, 1)

# --- 5. Geser (Translated) ---
# Menggeser citra 150 piksel ke kanan dan 105 piksel ke bawah
tx, ty = 150, 105
M_trans = np.float32([[1, 0, tx], [0, 1, ty]])
img_translated = cv2.warpAffine(img_rgb, M_trans, (cols, rows))
```

### Hasil Citra

```python
#202331249_Muhamad Nur Khanib
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
ax = axs.ravel()

# Tampilkan Citra Asli
ax[0].imshow(img_rgb)
ax[0].set_title('Citra Asli')
ax[0].axis('on') 

# Tampilkan Citra Rotated
ax[1].imshow(img_rotated)
ax[1].set_title('After Rotated')
ax[1].axis('on')

# Tampilkan Citra Resized
ax[2].imshow(img_resized)
ax[2].set_title('After Resized')
ax[2].axis('on')

# Tampilkan Citra Cropped
ax[3].imshow(img_cropped)
ax[3].set_title('After Cropped')
ax[3].axis('on')

# Tampilkan Citra Flipped
ax[4].imshow(img_flipped)
ax[4].set_title('After Flipped')
ax[4].axis('on')

# Tampilkan Citra Translated
ax[5].imshow(img_translated)
ax[5].set_title('After Translated')
ax[5].axis('on')

plt.tight_layout()
plt.show()
```

# Kompresi

```python
img_original = cv2.imread('kompresi_citra.jpg')
img_original_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
size_original_in_memory = img_original_rgb.nbytes / 1024
#202331249_Muhamad Nur Khanib
```

### Kompresi Lossy JPEG (Quality=10)

```python
#202331249_Muhamad Nur Khanib
encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), 10]
result, img_compressed_encode = cv2.imencode('.jpg', img_original, encode_param)

# Dekode kembali citra terkompresi untuk ditampilkan
img_compressed_rgb = cv2.imdecode(img_compressed_encode, 1)
img_compressed_rgb = cv2.cvtColor(img_compressed_rgb, cv2.COLOR_BGR2RGB)

# ukuran citra terkompresi (KB)
size_compressed_kb = len(img_compressed_encode) / 1024
```

###  Kuantisasi Warna RGB (4 Level per Channel)

```python
#202331249_Muhamad Nur Khanib
bins = 256 // 4 # Jumlah interval untuk setiap level
img_quantized_rgb = np.floor(img_original_rgb / bins) * bins + bins / 2
img_quantized_rgb = img_quantized_rgb.astype(np.uint8)

# Dapatkan ukuran citra kuantisasi dalam memori (KB)
size_quantized_in_memory = img_quantized_rgb.nbytes / 1024
```

### Hasil Citra

```python
#202331249_Muhamad Nur Khanib
fig, axs = plt.subplots(1, 3, figsize=(20, 7)) # Satu baris, tiga kolom
ax = axs.ravel()

# Citra Asli
ax[0].imshow(img_original_rgb)
ax[0].set_title(f'Asli\nUkuran: {size_original_in_memory:.2f} KB (In-Memory)')
ax[0].axis('off')

# Citra Hasil Kompresi Lossy
ax[1].imshow(img_compressed_rgb)
ax[1].set_title(f'Lossy JPEG Q=10\nUkuran: {size_compressed_kb:.2f} KB')
ax[1].axis('off')

# Citra Hasil Kuantisasi Warna
ax[2].imshow(img_quantized_rgb)
ax[2].set_title(f'Kuantisasi RGB (4 Level)\nUkuran: {size_quantized_in_memory:.2f} KB (In-Memory)')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
